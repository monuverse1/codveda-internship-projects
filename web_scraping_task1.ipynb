# Task 1: Data Collection and Web Scraping
This project scrapes quotes from the website quotes.toscrape.com using Python, BeautifulSoup, and requests. The extracted quotes are stored in a CSV file using pandas.

```python
# Import libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python

# URL of the website we want to scrape
url = "https://quotes.toscrape.com/"

# Send request to the website and store response
response = requests.get(url)

# Print HTML content of the webpage
print("Website request successful")
```

```python
# Parse the HTML content using html parser
soup = BeautifulSoup(response.text, "html.parser")
# Extract all quotes
quotes = soup.find_all("div", class_="quote")
# Print title of the webpage
print(soup.title)
```

```python
# Create empty list
data = []

# Extract quote and author
for quote in quotes:
    text = quote.find("span", class_="text").text
    author = quote.find("small", class_="author").text
    
    data.append({
        "Quote": text,
        "Author": author
    })

# Convert to DataFrame
df = pd.DataFrame(data)

# Save to CSV
df.to_csv("quotes_data.csv", index=False, encoding="utf-8")

# Show first rows
df.head(10)
```
